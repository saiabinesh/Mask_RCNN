```python
import urllib.request, urllib.error, urllib.parse, base64
from PIL import Image,ImageTk
import io
import tkinter, os, time

#File names
ROOT_DIR = os.path.abspath("../")
# Directory of images to run detection on
IMAGE_DIR = os.path.join(ROOT_DIR, "images")
file_names = os.listdir(IMAGE_DIR)

root = tkinter.Tk()
label = tkinter.Label(root)
label.pack()
img = None
tkimg = [None]  # This, or something like it, is necessary because 
# if you do not keep a reference to PhotoImage instances, they get garbage collected.

delay = 2000   # in milliseconds
def loopCapture(image_path):
    print("capturing")
    tkimg[0] = ImageTk.PhotoImage(Image.open(image_path))
    label.config(image=tkimg[0])
    root.update_idletasks()
    root.after(delay, loopCapture)
    print(delay)

for image_name in file_names:
    #time.sleep(10)
    print(image_name)
    loopCapture((os.path.join(IMAGE_DIR, image_name)))

#loopCapture('masked_HighresScreenshot00071.png')
root.mainloop()
```
